```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.

def modify_values(num, my_list):
    """Function to show how mutable and immutable arguments behave."""

    # Trying to modify immutable argument (integer)
    num += 10  # This creates a new `num` instead of modifying the original

    # Modifying mutable argument (list)
    my_list.append(99)  # This modifies the original list

    print("Inside function:")
    print("num (immutable) =", num)
    print("my_list (mutable) =", my_list)

# Immutable variable (integer)
x = 5
# Mutable variable (list)
my_list = [1, 2, 3]

print("Before function call:")
print("x (immutable) =", x)
print("my_list (mutable) =", my_list)

# Calling function
modify_values(x, my_list)

print("\nAfter function call:")
print("x (immutable) =", x)  # Unchanged
print("my_list (mutable) =", my_list)  # Changed!


```

```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types.


# Taking user input
user_input = input("Enter a value: ")


try:
    int_value = int(user_input)
except ValueError:
    int_value = "Cannot convert to integer"


try:
    float_value = float(user_input)
except ValueError:
    float_value = "Cannot convert to float"


bool_value = bool(user_input)

# Displaying results
print("\nConversion Results:")
print(f"String: '{user_input}'")
print(f"Integer: {int_value}")
print(f"Float: {float_value}")
print(f"Boolean: {bool_value}")




```

```python
# 10. 0. Write code to demonstrate type casting with list elements


string_list = ["10", "20.5", "0", "45", "", "100"]


int_list = [int(x) if x.isdigit() else None for x in string_list]


float_list = [float(x) if x.replace('.', '', 1).isdigit() and x.count('.') <= 1 else None for x in string_list]


bool_list = [bool(x) for x in string_list]

print("Original List:", string_list)
print("Integer List: ", int_list)
print("Float List:   ", float_list)
print("Boolean List: ", bool_list)



```

```python
# 14. 4. Write a program to reverse a string using a while loop.


# Take user input
original_string = input("Enter a string: ")


reversed_string = ""
index = len(original_string) - 1


while index >= 0:
    reversed_string += original_string[index]
    index -= 1


print("Reversed String:", reversed_string)
```

```python
# 15.  Write a Python program to calculate the factorial of a number provided by the user using a while loop.

# take user input
num = int(input("Enter a number: "))


factorial = 1


if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("Factorial of 0 is 1.")
else:
    i = num
    while i > 0:
        factorial *= i
        i -= 1

    print(f"Factorial of {num} is {factorial}.")



```
